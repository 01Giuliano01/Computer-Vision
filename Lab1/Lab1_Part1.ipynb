## 3) Artificial neuron — from math to code

A neuron computes:

$$
z = \sum_i w_i x_i + b
$$

Then applies an activation function:

$$
y = g(z)
$$

This section connects directly to the theory seen in class.


```python
x = torch.tensor([1.0, -2.0, 3.0])
w = torch.tensor([0.2, 0.4, -0.1])
b = torch.tensor(0.1)

z = torch.sum(x * w) + b
z
```

### Activation functions

1. Implement **ReLU**
2. Implement **Sigmoid**
3. Apply both to `z` and compare the outputs

Which activation preserves negative values?

None of them preserves negative values.

ReLU sets negative values to 0, and Sigmoid maps them to values between 0 and 1, so both remove negative values.

```python
# TODO
def relu(z):
  return torch.max(z, torch.zeros_like(z))

# TODO

def sigmoid(z):
  return 1 / (1 + torch.exp(-z))

y_relu = relu(z)
y_sigmoid = sigmoid(z)

y_relu, y_sigmoid

```

## 4) Autograd and gradients

PyTorch uses **automatic differentiation** to compute gradients
using the **chain rule** (backpropagation).


```python
x = torch.tensor([1.0, 2.0, -1.0], requires_grad=True)
w = torch.tensor([0.5, -0.3, 0.8], requires_grad=True)
b = torch.tensor(0.2, requires_grad=True)

z = torch.sum(x * w) + b
loss = (z - 1.0) ** 2

loss.backward()

print("loss:", loss.item())
print("grad w:", w.grad)
print("grad b:", b.grad)
```

### 🔍 Conceptual question

- If `b.grad > 0`, should `b` increase or decrease after a gradient descent step?
Explain **why** in one sentence.

If b.grad > 0, then b should decrease after a gradient descent step, because gradient descent updates parameters by subtracting the gradient to move in the direction that reduces the loss.



## 5) Toy classification dataset

We create a **linearly separable** dataset.

Label rule:
- class = 1 if `x₁ + x₂ + x₃ > 0`
- else class = 0

This mimics a very simple classification problem.


```python
# TODO: generate a dataset of size N=500 with 3 features
X = torch.randn(500, 3)

# TODO: compute labels
y = (X.sum(dim=1) > 0).float().unsqueeze(1) # shape (N, 1)

# TODO: split into train (80%) and validation (20%)
split = int(0.8 * 500)
X_train, X_val = X[:split], X[split:]
y_train, y_val = y[:split], y[split:]
```

## 6) Model definition

We define a small **MLP** (fully-connected network):

`3 → 16 → 8 → 1`

Activation: ReLU  
Output: raw logits (no sigmoid)


```python
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(3, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, 1)
        )

    def forward(self, x):
        return self.net(x)

# TODO: create model and move it to the GPU
model = MLP().to(device)

print("Model on:", device)
```

###  parameters

1. Compute **by hand** the total number of parameters
2. Verify your answer using PyTorch


```python
# TODO: count parameters with PyTorch
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

total_params

```

## 7) Training loop

You must complete the full training loop:
- forward pass
- loss computation
- backward pass
- optimizer step

Loss: `BCEWithLogitsLoss`
Optimizer: `SGD`


```python
# TODO: move data to device
X_train_d = X_train.to(device)
y_train_d = y_train.to(device)
X_val_d = X_val.to(device)
y_val_d = y_val.to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

for epoch in range(20):
    model.train()
    optimizer.zero_grad()

    # TODO: forward
    logit = model(X_train_d)

    # TODO: loss
    loss = criterion(logit, y_train_d)

    # TODO: backward
    loss.backward()

    # TODO: update
    optimizer.step()

    if epoch % 5 == 0:
        print("Epoch", epoch, "| loss =", float(loss))
```

## 8) Evaluation

1. Apply `sigmoid` to the logits
2. Convert probabilities to predictions
3. Compute **accuracy** on the validation set


```python
# TODO: evaluation
model.eval()
with torch.no_grad():

     logits = model(X_val_d)
     probs = torch.sigmoid(logits)
     preds = (probs > 0.5).float()
     accuracy = (preds == y_val_d).float().mean()
     print("Accuracy:", accuracy.item())
```

## 9) Reflection questions (answer inside the markdown)

1. Why do we **not** apply sigmoid inside the model?

We use BCEWithLogitsLoss, which already applies the sigmoid function internally in a numerically stable way. Adding sigmoid inside the model would apply it twice and can cause training instability.


2. What would happen if we removed all ReLU activations?

Without ReLU, the network becomes a stack of linear layers, which is equivalent to a single linear model. This would reduce the model’s capacity and prevent it from learning non-linear patterns.

3. How does this toy problem relate to image classification?

This toy problem follows the same pipeline as image classification: data → neural network → loss → optimization. The main difference is that images have many more features (pixels) and require deeper, more complex models.
